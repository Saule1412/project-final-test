```python
import pandas as pd

audience_df = pd.read_excel('test_data.xlsx', sheet_name='Данные об аудитории')
audience_df.head()
```

```python
mau = audience_df['user_id'].nunique()
print("MAU продукта равен:", mau)
```

```python
dau_by_day = audience_df.groupby('date')['user_id'].nunique()

mean_dau = dau_by_day.mean()
print("Средний DAU продукта равен:", mean_dau)
```

```python
users_nov1 = audience_df[audience_df['date'] == '2023-11-01']['user_id'].unique()

users_nov2 = audience_df[audience_df['date'] == '2023-11-02']['user_id'].unique()

returned = len(set(users_nov1).intersection(set(users_nov2)))

retention_1d = (returned / len(users_nov1)) * 100

print(f"Пользователей 1 ноября: {len(users_nov1)}")
print(f"Вернулись 2 ноября: {returned}")
print(f"Retention 1-го дня: {retention_1d:.1f}%")
```

```python
users_with_views = audience_df[audience_df['view_adverts'] > 0]

unique_users_with_views = users_with_views['user_id'].nunique()

conversion = (unique_users_with_views / 7639) * 100

print(f"Посмотрели объявления: {unique_users_with_views} пользователей")
print(f"Конверсия: {conversion:.1f}%")
```

```python
views_per_user = audience_df.groupby('user_id')['view_adverts'].sum()

mean_views = views_per_user.mean()

print(f"Среднее количество просмотров на пользователя: {mean_views:.1f}")
```

```python
ab_df = pd.read_excel('test_data.xlsx', sheet_name='Данные АБ тестов')
ab_df.head()
```

```python
import scipy.stats as stats

for i in[1, 2, 3]:

    exp_data = ab_df[ab_df['experiment_num'] == i]

    control_rev = exp_data[exp_data['experiment_group'] == 'control']['revenue']
    test_rev = exp_data[exp_data['experiment_group'] == 'test']['revenue']
    
    arpu_control = control_rev.mean()
    arpu_test = test_rev.mean()
    
    t_stat, p_val = stats.ttest_ind(control_rev, test_rev, equal_var=False)
    
    print(f"--- Эксперимент №{i} ---")
    print(f"ARPU Контроль: {arpu_control:.2f} | ARPU Тест: {arpu_test:.2f}")
    print(f"p-value: {p_val:.5f}")
    if p_val < 0.05:
        print("Результат: СТАТИСТИЧЕСКИ ЗНАЧИМ! (Есть реальная разница)\n")
    else:
        print("Результат: Незначим (Разница случайна)\n")
```
